crear un DF con los casos de homicidios

```python
# mostrar los casos donde veh acusado es diferente al veh fallecido ya que correponde a los casos de homicidio
df_homicidios = df_hyv[df_hyv['VEH_ACUSADO'] != df_hyv['VEH_FALLECIDO']]
df_homicidios
```

```python
df_homicidios['VEH_ACUSADO'].unique()
```

```python
acusados = ['OBJETO FIJO', 'PASAJEROS', 'CARGAS', 'MOTO', 'MULTIPLE','SD','OTRO','BICICLETA']
```

Eliminar del registro aquellos acusados quer por razones no logicas puedan ser considerados imputados de homicidio

```python
acusados = ['OBJETO FIJO', 'PASAJEROS','SD','OTRO']

df_homicidios[~df_homicidios['VEH_ACUSADO'].isin(acusados)].sample(frac=1)
```

```python
df_homicidios.info()
```

```python
# Crear una columna para el semestre (1 para enero a junio, 2 para julio a diciembre)
df_homicidios['SEMESTRE'] = df_homicidios['MES_F'].apply(lambda x: 1 if x <= 6 else 2)
df_homicidios
```

************************* crear metricas KPI 1 ********************

Filtrar solo los casos del segundo semestre de cada año

```python
df_homicidios_kpi_1 = df_homicidios[df_homicidios['SEMESTRE']==2]
df_homicidios_kpi_1
```

"Definimos a la tasa de homicidios en siniestros viales como el número de víctimas fatales en accidentes de tránsito por cada 100,000 habitantes en un área geográfica durante un período de tiempo específico. Su fórmula es: (Número de homicidios en siniestros viales / Población total) * 100,000"

```python
pobl_total = 3120612
#fuente = https://www.argentina.gob.ar/caba#:~:text=Poblaci%C3%B3n%3A%203.120.612%20habitantes%20(Censo%202022).
```

```python
nro_homicidios = df_homicidios_kpi_1['VEH_FALLECIDO'].count()
nro_homicidios
```

Formula 'tasa de homicidios en siniestros viales'

```python
tasa_hom_siniestros=((nro_homicidios/pobl_total)*100000).round(2)
print(tasa_hom_siniestros,'%')
```

```python
# cantidad de fallecidos por semestres en el año 2020 y 2021
segundo_semestre = df_homicidios_kpi_1.groupby(['AÑO_F','SEMESTRE'])['VEH_FALLECIDO'].count()
segundo_semestre
```

```python
# cantidad de fallecidos por semestres en el año 2020 y 2021 por sexo
segundo_semestre_sexo = df_homicidios_kpi_1.groupby(['AÑO_F','SEMESTRE','SEXO'])['VEH_FALLECIDO'].count()
segundo_semestre_sexo
```

```python
# cantidad de fallecidos por semestres en el año 2020 y 2021 por localidad
segundo_semestre_localidad = df_homicidios_kpi_1.groupby(['AÑO_F','SEMESTRE','TIPO_DE_CALLE'])['VEH_FALLECIDO'].count()
segundo_semestre_localidad
```

```python
# cantidad de fallecidos por semestres en el año 2020 y 2021 por edad
segundo_semestre_edad = df_homicidios_kpi_1.groupby(['AÑO_F','SEMESTRE','EDAD'])['VEH_FALLECIDO'].count()
segundo_semestre_edad
```

```python
# cantidad de fallecidos por semestres en el año 2020 y 2021 por responsable
segundo_semestre_responsable = df_homicidios_kpi_1.groupby(['AÑO_F','SEMESTRE','VEH_ACUSADO'])['VEH_FALLECIDO'].count()
segundo_semestre_responsable
```

```python
# cantidad de fallecidos por semestres en el año 2020 y 2021 por responsable y edad
segundo_semestre_responsable_edad = df_homicidios_kpi_1.groupby(['AÑO_F','SEMESTRE','VEH_ACUSADO','EDAD'])['VEH_FALLECIDO'].count()
segundo_semestre_responsable_edad
```

```python
# cantidad de fallecidos por semestres en el año 2020 y 2021 por responsable y sexo
segundo_semestre_responsable_sexo = df_homicidios_kpi_1.groupby(['AÑO_F','SEMESTRE','SEXO','VEH_ACUSADO'])['VEH_FALLECIDO'].count()
segundo_semestre_responsable_sexo
```
